```python
# 1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def find_sum(*nums):
    total=0
    for i in nums:
        if(i%2==0):
            total+=i
    return total

a=find_sum(10,67,34,98,12,-456,0,34)
print(a)  
print("-------------------or")
num=[10,67,34,98,12,-456,0,34]
sum(i for i in num if i%2==0)
```

```python
# 4.Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime1(num):
    if num<=1:
        a=f"{num} is not prime"
        return a
    for i in range(2,num//2+1):
        if (num%i==0):
            return f"{num} is not prime"
    return f"{num} is prime"
num=int(input("Enter any num between 1 and 201"))  
print(prime1(num))
```

```python
# 5.# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibo(term):
    a=0
    b=1
    print("fibonacci series:")
    for i in range(term):
        yield a
        a,b=b,a+b
    

a=fibo(10)
for i in range(10):
    print(next(a))
```

```python
# 6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_2(num):
    for i in range(num+1):
        yield 2**i

exponent=10
b=power_2(exponent)
for i in range(exponent+1):
    print(next(b))
```

```python
# 7.# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(filename):
    with open(filename, "r") as f:
        for line in f:
            yield line

for line in read_file("sample.txt"):
    print(line, end="")
```

```python
5. What are iterators in Python and how do they differ from iterables?

Iterable: An object that can be looped over like tuple,list,string.

Iterator: An object that produces values one at a time using next().

Example:
nums = [1, 2, 3]   # iterable
it = iter(nums)    # iterator
print(next(it))
print(next(it))
```

```python
6. Explain the concept of generators in Python and how they are defined.

Generators are special functions that use yield to produce values one at a time.

Example:
def count():
    yield 1
    yield 2
    yield 3

for i in count():
    print(i)
```

```python
9. Explain the purpose and usage of the map() function in Python.

map() applies a function to every element of an iterable.

Example:
nums = [1, 2, 3, 4]

result = list(map(lambda x: x*x, nums))

print(result)

Output:

[1, 4, 9, 16]
```

```python
# 10. What is the difference between map(), reduce(), and filter()?
# Functions
# map():Transform each item	
# filter():Select items based on condition	
# reduce():Combine items into a single value
# Example:


nums = [1, 2, 3, 4]

print(list(map(lambda x: x*2, nums)))
print(list(filter(lambda x: x%2==0, nums)))
print(reduce(lambda a,b: a+b, nums))

```
